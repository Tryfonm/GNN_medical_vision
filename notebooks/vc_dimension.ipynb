```python
from scipy.optimize import fsolve
import numpy as np


def equation(N):
    return 80 * np.log(4 * (2 * N + 1) / 0.05) - N


# Using fsolve to find the root of the equation
solution = fsolve(equation, 10000)  # Starting from N = 10
solution, equation(solution)
```
